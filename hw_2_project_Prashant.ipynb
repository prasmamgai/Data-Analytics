```python
import pandas as pd
from scipy.stats import ttest_ind, chi2_contingency, pearsonr
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customers = pd.read_csv(r"E:\Data Internship\Datasets\customers.csv")
transactions = pd.read_csv(r"E:\Data Internship\Datasets\transaction.csv")
web_sessions = pd.read_csv(r"E:\Data Internship\Datasets\web_sessions.csv")
campaigns = pd.read_csv(r"E:\Data Internship\Datasets\marketing_campaigns.csv")
```

### Specific questions

Q1. Do customers from the UK spend more per transaction than customers from Germany?

```python
merged = transactions.merge(customers, on="customer_id")

uk = merged[merged["country"] == "UK"]["purchase_value"]
germany = merged[merged["country"] == "Germany"]["purchase_value"]

t_stat, p_value = ttest_ind(uk, germany, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("There is a significant difference in purchase values between UK and Germany.")
else:
    print("There is no significant difference in purchase values between UK and Germany.")
```

Q2. Do discounted purchases differ in average value from non-discounted purchases?

```python
discounted = transactions[transactions["discount_used"] == 1]["purchase_value"]
non_discounted = transactions[transactions["discount_used"] == 0]["purchase_value"]

t_stat, p_value = ttest_ind(discounted, non_discounted, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("There is a significant difference in purchase values between discounted and non-discounted transactions.")
else:
    print("There is no significant difference in purchase values between discounted and non-discounted transactions.")
```

Q3. Do male and female customers differ in average purchase value?

```python
merged = transactions.merge(customers, on="customer_id")

male = merged[merged["gender"] == "Male"]["purchase_value"]
female = merged[merged["gender"] == "Female"]["purchase_value"]

t_stat, p_value = ttest_ind(male, female, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("There is a significant difference in purchase values between males and females.")
else:
    print("There is no significant difference in purchase values between males and females.")
```

Q4. Do customers acquired via email spend more than those from social channels?

```python
merged = transactions.merge(customers, on="customer_id")

email = merged[merged["acquisition_channel"] == "email"]["purchase_value"]
social = merged[merged["acquisition_channel"] == "social"]["purchase_value"]

t_stat, p_value = ttest_ind(email, social, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("There is a significant difference in customers acquired via email more than social channels.")
else:
    print("There is no significant difference in customers acquired via email more than social channels.")
```

Q5. Is device type associated with conversion?

```python
cont_table = pd.crosstab(web_sessions["device_type"],web_sessions["converted"])

chi2, p_value, dof, expected = chi2_contingency(cont_table)

print("Chi-square:", chi2)
print("P-value:", p_value)

if p_value < 0.05:
    print("There is a significant association between device type and conversion.")
else:    
    print("There is no significant association between device type and conversion.")
```

Q8. Is time on site correlated with purchase value?

```python
merged = web_sessions.merge(transactions.groupby("customer_id")["purchase_value"].mean().reset_index(),on="customer_id")

corr, p_value = pearsonr(merged["time_on_site"],merged["purchase_value"])

print("Correlation:", corr)
print("P-value:", p_value)

if p_value < 0.05:
    print("There is a significant correlation between time on site and purchase value.")
else:
    print("There is no significant correlation between time on site and purchase value.")
```

### Challenges

C1. Explore whether high-spending customers behave differently in terms of session time.

Define “high spending” yourself.

```python
avg_spend = transactions.groupby("customer_id")["purchase_value"].mean()

threshold = avg_spend.quantile(0.75)

high_spenders = avg_spend[avg_spend >= threshold].index
other_customers = avg_spend[avg_spend < threshold].index

high_time = web_sessions[web_sessions["customer_id"].isin(high_spenders)]["time_on_site"]

other_time = web_sessions[web_sessions["customer_id"].isin(other_customers)]["time_on_site"]

t_stat, p_value = ttest_ind(high_time, other_time, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("High spenders spend significantly more time on the site than other customers.")
else:
    print("There is no significant difference in time spent on the site between high spenders and other customers.")
```
